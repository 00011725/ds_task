```python
 g2p = G2p()
```

```python
# Phoneme craetion using g2p 
def transcribe_tokens(tokens):
    return ' '.join([' '.join(g2p(token)) for token in tokens])

def analyze_text_complexity(tokens):
    # Calculate average word length
    avg_word_length = np.mean([len(token) for token in tokens])
    
    # Calculate syllable count
    syllable_count = sum(count_syllables(token) for token in tokens)

    return avg_word_length, syllable_count

def count_syllables(word):
    # Basic syllable counting using regex
    word = word.lower()
    syllable_pattern = re.compile(r'[aeiouy]+')
    return len(syllable_pattern.findall(word))


# Process each row in the DataFrame
df['phonem'] = df['tokens'].apply(transcribe_tokens)
df['avg_word_length'], df['syllable_count'] = zip(*df['tokens'].apply(analyze_text_complexity))
print(df[['tokens', 'phonemes', 'avg_word_length', 'syllable_count']])
```
