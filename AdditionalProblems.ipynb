```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def normalDistribution(N, mu, sigma):
    samples = np.random.normal(mu, sigma, N)# N -number of samples, mu- mean, sigma -st dev
    plt.figure(figsize=(10, 4))
    sns.histplot(samples, bins=30, color = 'green') 
    plt.title(f'Histogram of {N} Samples from Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
N = 1000 
normalDistribution(N, mu=0, sigma=1)
```

```python
import pandas as pd

#Load the dataset
df = pd.read_csv('C:/Users/User/Desktop/DataScience/datasets/student-scores.csv')
df.head()
```

```python
buckets = ['<50', '<75', '<90', '<100']  

all_scores = ['math_score', 'history_score', 'physics_score', 
            'chemistry_score', 'biology_score', 'english_score', 
            'geography_score']

def checkAllScores(row):
    scores = row[all_scores]
    if all(score == 100 for score in scores):
        return '=100'
    elif all(score > 90 for score in scores):
        return '>90'
    elif all(score > 75 for score in scores):
        return '>75'
    elif all(score > 50 for score in scores):
        return '>50'
    else:
        return '<=50'

df['score_bucket'] = df.apply(checkAllScores, axis=1)
print(df[['id', 'first_name', 'last_name','score_bucket']])
```
